```python
import pandas as pd
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, DatetimeTickFormatter
from bokeh.models.tools import HoverTool
from bokeh.models.widgets import Tabs, Panel
```

```python
#Configure data
data = pd.read_hdf('Case003.sum')
#print(data)
del data['CVP2']
del data['PVC']

source = ColumnDataSource(data)
#data.columns.tolist()
```

```python
#HR
HR = figure()
HR.line(x='DateTime', y='HR',
         source=source,
         color = 'purple',
         line_width = 2)

HR.title.text = 'Relation of HR and Time'
HR.xaxis.axis_label = 'Time'
HR.yaxis.axis_label = 'Heart rate'

HR.xaxis.formatter=DatetimeTickFormatter(
        hours=["%d %b %Y %H:%M:%S"],
        days=["%d %b %Y %H:%M:%S"],
        months=["%d %b %Y %H:%M:%S"],
        years=["%d %b %Y %H:%M:%S"],
        minutes=["%H:%M:%S"],
        seconds=["%H:%M:%S:%f"],
    )

hover = HoverTool()
hover.tooltips=[
    ('Heart Rate', '@HR'),
    ('DateTime', '@DateTime{%F %H:%M:%S}')
]
hover.formatters = { "DateTime": "datetime"}

HR.add_tools(hover)
```

```python
tabs = Tabs(tabs=[HR_panel, SPO2_panel, NBPS_panel, NBPD_panel, NBPM_panel, AR1M_panel])
show(tabs)
```
