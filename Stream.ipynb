```python
i = 0

def modify_doc(doc):
    def update():
        rollover = 100
        global i
        
        #Construct the new values for all columns, and pass to stream
        new_data1 = dict(x=[data.index[i]],y=[data['HR'][i]])
        new_data2 = dict(x=[data.index[i]],y=[data['SPO2-%'][i]])
        new_data3 = dict(x=[data.index[i]],y=[data['NBP-S'][i]])
        new_data4 = dict(x=[data.index[i]],y=[data['NBP-D'][i]])
        new_data5 = dict(x=[data.index[i]],y=[data['NBP-M'][i]])
        new_data6 = dict(x=[data.index[i]],y=[data['AR1-M'][i]])
        
        #Update live value
        p1.text = str(data["HR"][i])
        p2.text = str(data["SPO2-%"][i])
        p3.text = str(data["NBP-S"][i])
        p4.text = str(data["NBP-D"][i])
        p5.text = str(data["NBP-M"][i])
        p6.text = str(data["AR1-M"][i])
        
        #Scroll live values in waveform view and prevents overflow
        source1.stream(new_data1, rollover) 
        source2.stream(new_data2, rollover)
        source3.stream(new_data3, rollover)
        source4.stream(new_data4, rollover)
        source5.stream(new_data5, rollover)
        source6.stream(new_data6, rollover)
        
        #Trend
        sourceT1.stream(new_data1) 
        sourceT2.stream(new_data2)
        sourceT3.stream(new_data3)
        sourceT4.stream(new_data4)
        sourceT5.stream(new_data5)
        sourceT6.stream(new_data6)
        
        i += 1
        
        # if there is no more data then remove the period_callback from the doc to stop updating the plot
        # this will only work when the dataframe is loaded all at once
        # this won't be practical when working with waveform data since you will need to read the files 
        # incrementally... still, the basic idea still applies

        if i > len(data) - 1:
            doc.remove_periodic_callback(loop_cb)
            return

    loop_cb = doc.add_periodic_callback(update, 100)
    doc.add_root(layouts)
```
